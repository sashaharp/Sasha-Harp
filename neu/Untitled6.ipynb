```python
import numpy as np
import matplotlib.pyplot as plt

xn = np.linspace(-2, 0, 200)
yn = np.linspace(-2, 0, 200)
x, y = np.meshgrid(xn,yn)

h = x**3 - y**3 - 3*x*y
b = abs(h) < 0.003 #altern zur nst berechnung bzw zur parametr

xn2 = np.linspace(0, 2, 200)
yn2 = np.linspace(0, 2, 200)
x2, y2 = np.meshgrid(xn2,yn2)

h2 = x2**3 - y2**3 - 3*x2*y2
b2 = abs(h2) < 0.003 #altern zur nst berechnung bzw zur parametr

xn3 = np.linspace(-2, 0, 200)
yn3 = np.linspace(0, 2, 200)
x3, y3 = np.meshgrid(xn3,yn3)

h3 = x3**3 - y3**3 - 3*x3*y3
b3 = abs(h3) < 0.003 #altern zur nst berechnung bzw zur parametr
b31 = y3[b3]>-1*x3[b3]
b32 = y3[b3]<-1*x3[b3]

plt.plot(x[b], y[b])
plt.plot(x2[b2], y2[b2])
plt.plot(x3[b3][b31], y3[b3][b31])
plt.plot(x3[b3][b32], y3[b3][b32])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

xs = np.linspace(-7, 7, 70)
ys = np.linspace(-7, 7, 70)
zs = np.linspace(0, 2, 70)
xs, ys, zs = np.meshgrid(xs, ys, zs)
xs = xs.flatten()
ys = ys.flatten()
zs = zs.flatten()
```

```python
bs = np.logical_and(xs**2+ys**2>1/2*(1+zs**2), xs**2+ys**2<2*(1+zs**2))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xs[bs], ys[bs], zs[bs], c=zs[bs], marker="o")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from random import random
```

```python
x = np.arange(400)
y = [0]
weight = random()
for xx in range(len(x)-1):
    if random() < weight:
        y.append(y[len(y)-1]+1)
    else:
        y.append(y[len(y)-1]-1)
    if random() < 0.3:
        weight = random()
y = np.array(y)
```

```python
plt.plot(x, y)
```

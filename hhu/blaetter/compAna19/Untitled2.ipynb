```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1 + 0j, 1 + 0j, 100)
f = x**(4/7)
g = (x**4)**(1/7)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.plot(x, f)
ax1.plot(x, np.imag(f))
ax2.plot(x, g)
ax2.plot(x, np.imag(g))
```

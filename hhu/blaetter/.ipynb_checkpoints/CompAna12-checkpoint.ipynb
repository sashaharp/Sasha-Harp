```python
import matplotlib.pyplot as plt
import numpy as numpy
from sympy import *
#import matplotlib.image as mpimg
```

```python
X = numpy.linspace(-5, 5, 400)
Y = numpy.linspace(-1, 4, 200)
x, y = numpy.meshgrid(X, Y)
f = -1*numpy.cos(y)*numpy.cosh(x)-4*y
"""
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, f)
"""
```

```python
foriginal = f.copy()
f -= numpy.amin(f)
f /= numpy.amax(f)
f = numpy.array([[[f[i,j], 0, 1-f[i,j]] for j in range(len(f[i]))] for i in range(len(f))])
```

```python
plt.imshow(f, extent=[-5,5,4,-1])
```

```python
xs = S("x")
ys = S("y")
fs = -1*cos(ys)*cosh(xs)-4*ys
dfs = (diff(fs, xs), diff(fs, ys))
zeros = solve([Eq(dfs[0], 0), Eq(dfs[1], 0)])
zeros
```

```python
realzeros = []
for z in zeros:
    if z[xs].is_real and z[ys].is_real:
        realzeros.append([z[xs], z[ys]])
realzeros
```

```python
for z in realzeros: #f[(numpy.abs(X - N(z[0].simplify()))).argmin(), (numpy.abs(Y - N(z[0].simplify()))).argmin()] = [1, 1, 1]
    yi = (numpy.abs(Y - N(z[1].simplify()))).argmin()
    xi = (numpy.abs(X - N(z[0].simplify()))).argmin()
    f[yi-4:yi+4, xi-4:xi+4] = [1, 1, 1]
plt.imshow(f, extent=[-5,5,4,-1])
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, foriginal, facecolors=f)
```

```python
a = S("a")
A = numpy.array([[7, 37-8*a, 3, 33-8*a],[-1, -6+a, -3, -8+a],[-1, 2-a, -3, -1*a]])
A1 = Matrix(A[:,:3])
A2 = Matrix(A[:,1:])
print(A1.det().simplify())
print(A2.det().simplify())
(A1.det()+A2.det()).simplify()
```

```python
print(solve(Eq(A1.det().simplify(),0),a))
```

```python
M = [numpy.array([[1, 1, -1],[1, 1, 1],[-1, 1, 0]]), numpy.array([[1, 1, 1],[1, 1, 1],[1, 1, 2]]), numpy.array([[1, 1, 1],[1, 1, 1],[1, 1, 0]])]
for m in M:
    print(Matrix(m[:1, :1]).det(), Matrix(m[:2, :2]).det(), Matrix(m).det())
```

```python
for m in M:
    if(Matrix(m).eigenvals().keys()[0]>=0, Matrix(m).eigenvals().keys()[1]>=0, Matrix(m).eigenvals().keys()[2]>=0):
        print(m)
        print("ist semi positiv definit")
    elif(Matrix(m).eigenvals().keys()[0]<=0, Matrix(m).eigenvals().keys()[1]<=0, Matrix(m).eigenvals().keys()[2]<=0):
        print(m)
        print("ist semi negativ definit")
    else:
        print(m)
        print("ist indefinit")
```

```python
import numpy as np
import matplotlib.pyplot as plt

xn = np.array([0, np.pi/2, np.pi, 3/2 * np.pi, 2 * np.pi])
yn = np.array([0, 1, 0, -1, 0])

x = np.linspace(0, 2 *np.pi, 50) #c werte zwischen a und b (linear, dh gleiche abstande, space wegen menge)
y = np.interp(x, xn, yn)

plt.plot(xn, yn, "H") #H = hexagon
plt.plot(x, np.sin(x), label = "$\sin{(x)}$")
plt.plot(x, y, label = "interp sin")
plt.legend()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
z = x * y



b = x**4+y**4+z**4 + 1000*(x**4+y**4)*(x**4+z**4)*(z**4+y**4) == 10


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x[b], y[b], z[b], cmap=plt.cm.viridis, rstride=1, cstride=1, linewidth=2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x = np.linspace(-4, 4, 200)
y=(1-np.e**x)/(1+np.e**x)
plt.plot(x, y)
```

```python
p1 = x < -np.pi/2
p2 = np.abs(x) < np.pi/2
p3 = x > np.pi/2
y=np.tan(x)
plt.plot(x[p1], y[p1])
plt.plot(x[p2], y[p2])
plt.plot(x[p3], y[p3])
```

```python
x = np.linspace(-4, 4, 9001)

y = 1/((x-4)**2*(x**2-1))

b1 = x < -1
b2 = [(b > -1 and b < 1) for b in x]
b3 = x > 1

plt.plot(x[b1], y[b1])
plt.plot(x[b2], y[b2])
plt.plot(x[b3], y[b3])
plt.axis(ymin=-20, ymax=20)
```

```python
x = np.linspace(0, 1, 200)
f1 = np.array([0.5*(1+(k/1)) for k in x])
f2 = np.array([0.5*(f1[k]+(x[k]/f1[k])) for k in range(len(x))])
f3 = np.array([0.5*(f2[k]+(x[k]/f2[k])) for k in range(len(x))])
sr = np.sqrt(x)

plt.plot(x, f1, label="$f_1(x)$")
plt.plot(x, f2, label="$f_2(x)$")
plt.plot(x, f3, label="$f_3(x)$")
plt.plot(x, sr, label="$\sqrt{x}$")
plt.title("Babylonisches Wurzelziehen")
plt.legend();
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3, 3, 200)
y = np.array([[i**j for i in x] for j in range(6)])

plt.plot(x, y[0], "k", label="$y = 1$")
for k in range(1, 6):
    plt.plot(x, y[k], label="$y = x^" + str(k) + "$")
plt.legend()
plt.axis(ymin=-10, ymax=10)
plt.show()
```

```python
x = np.linspace(-np.pi/2, np.pi/2, 100)
y = np.sin(x)

plt.plot(x, y, "k", label="$\sin{x}$")
plt.legend()
plt.axis(ymin=-np.pi/2, ymax=np.pi/2)
plt.show()
```

```python

plt.plot(y, x, "r", label="$\sin^{-1}{x}$")
plt.legend()
plt.axis(ymin=-np.pi/2, ymax=np.pi/2)
plt.show()
```

```python
x1 = np.linspace(-1, 1, 200)[1:200]
x2 = np.linspace(-np.pi/2, np.pi/2, 200)
f, (ax1, ax2) = plt.subplots(1, 2)
ax1.plot( np.sin(x2), x2, "k", label="$\sin{x}$")
ax2.plot(x1, np.arcsin(x1), "r", label="$\sin^{-1}{x}$")
ax1.legend()
ax2.legend()
ax1.axis(ymin=-np.pi/2, ymax=np.pi/2, xmin=-1, xmax = 1)
ax2.axis(ymin=-np.pi/2, ymax=np.pi/2)
plt.show()
```

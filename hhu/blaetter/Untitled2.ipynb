```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

I = np.array([-3/2, -1/2, 1/2, 3/2])
J = np.array([-1, 0,1])

x = np.linspace(-3/2, 3/2, 60)
y = np.linspace(-1, 1, 60)
X, Y = np.meshgrid(x, y)

h = np.minimum(np.array([[np.sum(np.array([np.sum(1 / ((xn - i)**2 + (yn - J)**2)) for i in I])) for yn in y] for xn in x ]), 50)

b = h<=50

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
# 1x1-Matrix von Plots, davon Element 1
#ax.plot_surface(X, Y, Z, cmap=plt.cm.plasma)
ax.plot_surface(X, Y, h, cmap=plt.cm.coolwarm, alpha=.35, rstride=1, cstride=1, linewidth=2)
ax.set_zlim3d(bottom=0, top=50)
```

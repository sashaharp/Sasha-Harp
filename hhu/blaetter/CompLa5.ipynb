```python
import numpy as np
import matplotlib.pyplot as plt

xn = np.arange(3)
yn = -np.arange(4)
X, Y = np.meshgrid(xn, yn)
print(X)
print(Y)
print(X.dtype, ',', Y.dtype)
```

```python
xn = np.arange(3).reshape(1,-1)
yn = -np.arange(4).reshape(-1,1)
X=np.ones(4).reshape(-1, 1).astype(int)*xn
Y=np.ones(3).reshape(1, -1).astype(int)*yn
print(X)
print(Y)
assert(X.dtype==Y.dtype and Y.dtype=="int64")
```

```python
x = np.linspace(0, 2*np.pi, 50)
xn = np.arange(5)*np.pi/2
yn = np.array([0, 1, 0, -1, 0])
plt.plot(x, np.interp(x, xn, yn), label="interpolated $sin(x)$")
plt.plot(x, np.sin(x), label="")

plt.plot(xn, yn, "H")
plt.legend()
```

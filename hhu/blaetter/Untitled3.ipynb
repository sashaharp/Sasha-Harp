```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

t = np.linspace(-1, 100, 50000)
t2 = np.linspace(-100, -1, 50000)

x = -3*t/(1+t**3)
y = 3*t**2/(1+t**3)
x2 = -3*t2/(1+t2**3)
y2 = 3*t2**2/(1+t2**3)

b=x<=2

plt.plot((x[b]), (y[b]), "m")
plt.plot((x2), (y2), "m")
plt.axis(xmin=-2, xmax=2, ymin=-2, ymax=2)
plt.show()

```

```python
x = np.linspace(-2, 2, 50)
y = np.linspace(-2, 2, 50)
X, Y = np.meshgrid(x, y)
h = (X**3) - (Y**3) - (3*X*Y)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.contour(X, Y, h, [-2, -1, 0, 1, 2])
ax.plot_surface(X, Y, h, cmap=plt.cm.viridis, rstride=1, cstride=1, linewidth=2)
```
